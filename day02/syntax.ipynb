beautifulsoup 테스트

```python
from bs4 import BeautifulSoup
```

```python
html = '<h1 id="title">한빛출판네트워크</h1><div class="top"><ul class="menu"><li><a href=http://www.hanbit.co.kr/member/login.html class="login">로그인 </a></li></ul><ul class="brand"><li><a href="http://www.hanbit.co.kr/media/>한빛미디어<li><a href="http://www.hanbit.co.kr/academy/">한빛아카데미</a></li></ul></div>'

html
```

```python
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
print(soup.prettify())
```

```python
soup.h1 #특정한것만 볼 수 있음
```

```python
soup.div.ul.li.a #타고 들어가는거 가능
```

```python
soup.find("li") #여러개 중 하나만 찾음
```

```python
soup.find_all("li") #여러개 모두 찾음
```

```python
tag_a = soup.a
tag_a
```

```python
tag_a.attrs #속성(딕셔너리 형태)
```

```python
tag_a["href"]
```
