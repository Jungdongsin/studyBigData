### 히스토그램

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager
import matplotlib

```

```python
mydata = pd.read_csv("./data/mydata.csv", encoding="utf-8")
mydata
```

```python
#히스토그램
mydata[["HM_AVG"]].hist(bins=10)
```

```python
font_path = "C:/Windows/Fonts/malgun.ttf" # 맑은고딕 글자명
font_name = font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rc("font", family=font_name)
```

```python
plt.hist(mydata["HM_AVG"], bins=10)
plt.title("평균습도")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(mydata.HM_AVG,bins=10, kde=True)
```

```python
df = pd.DataFrame(mydata)
df.describe()
```

```python
import scipy as sp

sp.stats.describe(mydata)
```
